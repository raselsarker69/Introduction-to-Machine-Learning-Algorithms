# "Gradient descent with Linear regression 

```python
m = 0
c = 0
lr = 0.001   # learning rate.
n = len(x)

for i in range(500):
    yhat = (m*x) + c
    
    Dm = (-2/n)*sum(x*(y-yhat))
    Dc = (-2/n)*sum(y-yhat)
    
    #update the parameters
    m = m-(lr*Dm)
    c = c-(lr*Dc)
    
    # print(m,c)
print(f"Slope (m):", m) 
print(f"Intercept (c):", c) 
```
