```python
import pandas as pd 
import numpy as np  
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
    'Phone Model': ['Model A', 'Model B', 'Model C', 'Model D', 'Model E', 
                    'Model F', 'Model G', 'Model H', 'Model I', 'Model J'],
    'RAM (GB)': [4, 6, 8, 12, 3, 4, 8, 6, 4, 6],
    'ROM (GB)': [64, 128, 128, 256, 32, 64, 128, 64, 32, 128],
    'Camera (MP)': [12, 48, 64, 108, 8, 16, 32, 12, 13, 48],
    'Display Size (inches)': [6.1, 6.4, 6.7, 6.8, 5.8, 6.0, 6.5, 6.3, 5.5, 6.2],
    'Battery (mAh)': [3000, 4000, 4500, 5000, 2800, 3200, 4100, 3700, 3000, 4200],
    'Price (USD)': [399, 499, 599, 899, 299, 349, 549, 449, 299, 499]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['RAM (GB)', 'ROM (GB)', 'Camera (MP)', 'Display Size (inches)', 'Battery (mAh)']]
y = df['Price (USD)']
```

```python
X = X.values
y = y.values
```

```python
# Feature scaling (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Add bias term (column of 1's)
X_scaled = np.c_[np.ones(X_scaled.shape[0]), X_scaled]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Predict and evaluate
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
y_pred = reg.predict(X_test)
mse = mean_absolute_error(y_test, y_pred)
print("MAE:", mse)
```

# RMSE Calculations

```python
import math
math.sqrt(mse)
```

```python
plt.scatter(y_test, y_pred, color='red', marker='*', label = 'Mobile price predictions')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.legend()
plt.show()
```

```python
def predict_price(new_data, reg, scaler):
    new_data_scaled = scaler.transform([new_data])
    
    # Add bias term
    new_data_scaled = np.insert(new_data_scaled, 0, 1)
    
    # Predict price
    pred_price = reg.predict([new_data_scaled])
    return pred_price[0]


new_data = [16, 128, 12, 8.0, 7000]
predicted_price = predict_price(new_data, reg, scaler)
print("Predicted price for new mobile phone:", predicted_price)
```
